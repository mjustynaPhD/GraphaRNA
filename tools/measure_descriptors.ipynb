```python
import pickle
import os
from tqdm import tqdm
```

```python
data_path = "/home/mjustyna/RNA-GNN/data/RNA-PDB-clean/val-pkl"
files = os.listdir(data_path)
print(len(files))
```

```python
with open("all-output.txt", 'r') as f:
    lines = f.readlines()

lines = [l.split() for l in lines]
lines = [l for l in lines if len(l)>7 and len(l[1])>=12]
pairs = dict([(l[1], int(l[2])) for l in lines])
pairs
```

```python
residues_num = []
for f in tqdm(files):
    with open(f"{data_path}/{f}", 'rb') as pf:
        pick = pickle.load(pf)
        # print(pick.keys())
        n = pick['name'][:-4]
        if pairs[n] == 1:
            residues_num.append(sum(pick['c4_primes']))
    break
```

```python
import matplotlib.pyplot as plt
print(min(residues_num), max(residues_num))
plt.hist(residues_num)
```
